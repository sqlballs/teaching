# Linear Regression
# When we find features that have  data points placed linearly you may found a feature that we can use in a linear regression model.  Regression works on the line equation, y=mx+c.  The trend line is set through the data points to predict the outcome.  

# The variable we are predicting is referred to as Y.  The variable we are basing our predictions on is called the predictor variable and is referred to as X.  When there is only one predictor variable, the prediction method is called simple regression.  If there are multiple predictor varables are present then we call the prediction method multiple regression.

# We start by creating a simple linear regression.

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
#create a training data set that does not have the ID column, which is not useful, and the price, which we are trying to predict.
#convert dates in our core data set.  Data is from 2014, all 2014 homes flag as brand new=1, all others, flag as 0.
#create a new data set named labels, which contains all of the price data.
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
#print labels
labels
```

```python
#view the head information from train1
train1.head()
```

```python
#import the test_train_split module in order to spilt our labels and train1 data set into seperate data sets
from sklearn.cross_validation import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
#train our linear regression model
reg.fit(x_train,y_train)
```

```python
#observe the score four our linear regression model
reg.score(x_test,y_test)
```

# While this may be good enough, could it be better.

# Let's examine Gradient Boosting Regression.

# What is gradeint boosting regression?  It is a ML technique for regression and classification problems, in which a prediction model is produced based off the ensemble of weak prediction models, most typically decision trees.

# What are some reasons this could be effective? 
# 1. Data could be lost / forgotten by the system
# 2. The algorithm may be a weak learning system that needs a decision tree
# 3. We need to give the algorithm time to examine past patterns

```python
#import the libary from sklearn, create our variable where we define our boosting regressors
#n_estimator = the number of boosting stages to perform.  do not set it too high, it could overfit the model
#max_depth = the depth of decision tree nodes
#learning_rate = rate of learning the data
#loss = loss function to be optimized, ls refers to the least squares regression
#minimum sample split = number of sample to be split for learning the data
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
#train our model
clf.fit(x_train, y_train)
```

```python
#score our model
clf.score(x_test,y_test)
```

```python
#plot our training deviance
import numpy as np
params = {'n_estimators': 400, 'max_depth': 5, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'}
#compute test set deviance
t_sc = np.zeros((params['n_estimators']),dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)
```

```python
testsc = np.arange((params['n_estimators']))+1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, clf.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, t_sc, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')
```

```python
#Which features were most importaint in our model?
feature_importance = clf.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos,list(train1))
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```
